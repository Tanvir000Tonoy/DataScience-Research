```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Upload the Gym data in google drive
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Home Tasks/Week04/gym_membership.csv')
df = pd.DataFrame(df)
# Display last 5 rows
print(f'\nLast 5 rows \n \n  {df.tail()}')
# Display Info
print(df.info())
# Check data types
print(f'\nData types \n \n {df.dtypes}')
```

```python
# Task 1: Basic Matplotlib (3 plots)
# 1. Bar chart of gym membership types
# Use: gym['abonoment_type'].value_counts()
x = ['Standard','Premium']
y = df['abonoment_type'].value_counts()
plt.bar(x,y)
plt.title('Gym Membership Types')
plt.xlabel('Membership Type')
plt.ylabel('Count')
plt.show()

# 2. Histogram of time in gym
# Use: plt.hist(gym['avg_time_in_gym'])
plt.hist(df['avg_time_in_gym'])
plt.title('Time in Gym')
plt.xlabel('Time')
plt.ylabel('Count')
plt.show()
# 3. Scatter plot: Age vs visits per week
# Use: plt.scatter(gym['Age'], gym['visit_per_week'])
plt.scatter(df['Age'], df['visit_per_week'])
plt.title('Age vs Visits per Week')
plt.xlabel('Age')
plt.ylabel('Visits per Week')
plt.show()
```

```python
# 1. Box plot: Gender vs Age
# Use: sns.boxplot(data=gym, x='gender', y='Age')
sns.boxplot(data=df, x='gender', y='Age')
plt.title('Gender vs Age')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
# 2. Count plot: Days per week
# Use: sns.countplot(data=gym, x='days_per_week')
sns.countplot(data=df, x='days_per_week')
plt.title('Days per Week')
plt.xlabel('Days')
plt.ylabel('Count')
plt.show()
# 3. Bar plot: Average visits by membership type
# Use: sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
sns.barplot(data=df, x='abonoment_type', y='visit_per_week')
plt.title('Average Visits by Membership Type')
plt.xlabel('Membership Type')
plt.ylabel('Average Visits')
plt.show()
```
